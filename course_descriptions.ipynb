```python
# load osgoode_course_descriptions.json via pandas

import pandas as pd
import json
import html

with open("DATA/osgoode_course_descriptions.json", encoding="utf-8") as f:
    data = json.load(f)

# Combine all course lists into one DataFrame, with a column for the term
frames = []
for term, courses in data.items():
    temp = pd.DataFrame(courses)
    temp["term"] = term
    frames.append(temp)

df = pd.concat(frames, ignore_index=True)

df = df[["title", "instructor", "number", "term", "description"]]
df["description"] = df["description"].fillna("").apply(html.unescape)


```

```python
# Filter for courses related to AI, generative AI, ChatGPT, Copilot, or Claude
keywords = [
    "artificial intelligence", "generative ai", 
    "chatgpt", "copilot", "claude"
]

cols = ["title", "description"]
pattern = "|".join(keywords)

mask = df[cols].fillna("").apply(lambda c: c.str.lower().str.contains(pattern)).any(axis=1)
exclude = df["description"].fillna("").str.lower().str.contains("generative ai will be strictly prohibited")
ai_df = df[mask & ~exclude]
ai_df
```

```python
# Print the AI-related courses
for _, row in ai_df.iterrows():
    for col in ai_df.columns:
        print(f"{col.upper()}: {row[col]}")
    print("_________________________________")
```
